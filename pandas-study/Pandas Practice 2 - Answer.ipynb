```python
import pandas as pd
```

### Q1. `student2.csv` 파일을 불러와 `df` 변수에 저장해봅시다.

```python
df = pd.read_csv('student2.csv')
df.head()
```

### Q2. 데이터를 보니 잘못된 column이 하나가 포함되어 있네요. 그 column을 drop해봅시다.

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

### Q3. 출석 점수를 계산하여 `total_att`라는 column을 생성해봅시다.
* 출석 일수는 총 32일이다.
* 결석을 할 때마다 -1, 지각을 할 때마다 -0.5점이 감점이 된다.
* 감점이 -4점 이상이 되면 자동적으로 출석점수는 0점으로 된다.
* 출석 반영 비율은 10%이다.

```python
df['total_att'] = 10 - df['absent'] - df['late']*0.5
df.loc[df['total_att'] <= 6, 'total_att'] = 0
df.head()
```

### Q4. `total_ex_att`와 `total_att`로 총점을 구하여 `total` column을 생성해봅시다.

```python
df['total'] = df['total_ex_att'] + df['total_att']
df.head()
```

### Q5. `course_retake` column의 경우 Nan값이 존재합니다. 시스템 상 재수강을 하지 않는 학생들은 체크가 되지 않는 다고 하네요. Nan값을 `N`으로 채워봅시다.

```python
df['course_retake'] = df['course_retake'].fillna('N')
df.head()
```

### Q6. 학생들의 등수를 매겨서 `rank`라는 column을 생성해봅시다.
* 힌트 : `sort_values`를 이용해봅시다.
* 힌트 : `reset_index`를 사용해봅시다.

```python
df = df.sort_values(by=['total'], ascending=False)
df.head()
```

```python
df.reset_index(inplace=True)
df['rank'] = df.index + 1
df.drop(['index'], axis=1, inplace=True)
df.head()
```

### Q7. 학점을 매겨봅시다. 학점은 `grade`라는 column에 저장합니다.
* 5등까지는 `A+`를, 10등까지는 `A`를 줍니다.
* 15등까지는 `B+`를, 20등까지는 `B`를 줍니다.
* 그 외 등수는 `C+`를 부여합니다.

```python
def get_grade(rank):
    cat = ''
    if rank <= 5: cat = 'A+'
    elif rank <= 10: cat = 'A'
    elif rank <= 15: cat = 'B+'
    elif rank <= 20: cat = 'B'
    else: cat = 'C+'
    
    return cat

df['grade'] = df['rank'].apply(lambda x : get_grade(x))
df.head(100)
```

### Q8. 재수강 학생은 다른 학점을 부여해야한다고 합니다.
* 2016년 이상 학생은 B+까지 밖에 못 받습니다.
* 2015년 이하 학생은 A까지 밖에 못 받습니다.
* 비율의 변경이 있어도 교수님은 채우지 않습니다:(

```python
df.loc[((df['grade'] == 'A') | (df['grade'] == 'A+')) & (df['admin_year'] >= 2016) & (df['course_retake'] == 'Y'), 'grade'] = 'B+'
df.loc[(df['grade'] == 'A+') & (df['admin_year'] < 2016) & (df['course_retake'] == 'Y'), 'grade'] = 'A'

df.head(100)
```
