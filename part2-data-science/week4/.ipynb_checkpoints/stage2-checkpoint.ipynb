```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

## 1. 데이터 로드

```python
df = pd.read_csv('./data/iris.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

## 2. 데이터 시각화

### 2.1. 꽃받침의 길이와 너비

```python
sns.catplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', aspect=3)
```

### 2.2. 꽃잎의 길이와 너비

```python
sns.catplot(data=df, x='PetalLengthCm', y='PetalWidthCm', hue='Species', aspect=3)
```

### 2.3. Feature 간의 상관관계

```python
df_no_id = df.drop(['Id'], axis=1)
sns.pairplot(df_no_id, hue="Species")
plt.title("Iris Data의 Pair Plot")
plt.show()
```

### 2.4. 분포도
#### 2.3.1. 꽃받침의 길이

```python
facet = sns.FacetGrid(data=df, hue='Species', aspect=4)
facet.map(sns.kdeplot, 'SepalLengthCm')
facet.add_legend()

plt.show()
```

#### 2.3.2. 꽃받침의 너비

```python
sns.catplot(data=df, kind='violin', x='Species', y='SepalWidthCm', aspect=3)
plt.show()
```

## 3. 모델링

```python
input_data = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
target_data = df['Species']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split(input_data, target_data)

x_train.shape, x_valid.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=100)
forest.fit(x_train, y_train)
```

```python
print('training set accuracy :', forest.score(x_train, y_train))
print('valid set accuracy :', forest.score(x_valid, y_valid))
```

## 4. 예측하기

```python
forest.predict(x_valid)
```
