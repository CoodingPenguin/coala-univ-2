# stage1

```python
import pandas as pd 
import numpy as np
```

```python
df_train = pd.read_csv('./data/train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('./data/test.csv')
df_test.head()
```

## Age의 빈칸 채우기

```python
df_train.isna().sum()
```

```python
df_train['Age'].fillna(df_train['Age'].mean(), inplace=True)
```

```python
df_train.head()
```

```python
df_test['Age'].fillna(df_test['Age'].mean(), inplace=True)
df_test.head()
```

## Age를 세대별로 묶기

```python
df_train.loc[df_train['Age'] < 10, 'Age'] = 0
df_train.loc[(df_train['Age'] >= 10) & (df_train['Age'] < 20), 'Age'] = 1
df_train.loc[(df_train['Age'] >= 20) & (df_train['Age'] < 30), 'Age'] = 2
df_train.loc[(df_train['Age'] >= 30) & (df_train['Age'] < 40), 'Age'] = 3
df_train.loc[(df_train['Age'] >= 40) & (df_train['Age'] < 50), 'Age'] = 4
df_train.loc[df_train['Age'] > 50, 'Age'] = 5

df_train.head()
```

```python
df_test.loc[df_test['Age'] < 10, 'Age'] = 0
df_test.loc[(df_test['Age'] >= 10) & (df_test['Age'] < 20), 'Age'] = 1
df_test.loc[(df_test['Age'] >= 20) & (df_test['Age'] < 30), 'Age'] = 2
df_test.loc[(df_test['Age'] >= 30) & (df_test['Age'] < 40), 'Age'] = 3
df_test.loc[(df_test['Age'] >= 40) & (df_test['Age'] < 50), 'Age'] = 4
df_test.loc[df_test['Age'] > 50, 'Age'] = 5

df_test.head()
```

## Family Size 만들기

```python
# 형재자매수 + 부모님수  = 가족수
df_train['FamilySize'] = df_train['SibSp'] + df_train['Parch']
df_train.head()
```

```python
# 형재자매수 + 부모님수  = 가족수
df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch']
df_test.head()
```

## train, test 묶기 1차

```python
train = df_train[['Survived', 'Sex', 'Age', 'FamilySize']]
test = df_test[['Sex', 'Age', 'FamilySize']]
```

```python
train.head()
```

## Fare 빈칸을 채우기

```python
print(df_train.isna().sum())
print('='*50)
print(df_test.isna().sum())
```

```python
df_test['Fare'].fillna(df_test['Fare'].mean(), inplace=True)
```

```python
df_train['Embarked'].value_counts()
```

```python
df_train['Sex'].value_counts()
```

## Embarked 빈칸 채우기

```python
df_train['Embarked'].fillna('S', inplace=True)
df_train.head()
```

```python
df_test['Embarked'].fillna('S', inplace=True)
df_test.head()
```

```python
print(df_train.isna().sum())
print('='*50)
print(df_test.isna().sum())
```

## train, test 묶기 2차

```python
train = df_train[['Survived', 'Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]
test = df_test[['Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]

train.head()
```

## Challenge 1

```python
# 성별
binary_sex = {'male':0, 'female':1}
df_train['Sex'] = df_train['Sex'].map(binary_sex)
df_test['Sex'] = df_test['Sex'].map(binary_sex)

df_train.head()
```

```python
# 선착장
binary_embarked = {'S':0, 'C':1, 'Q':2}
df_train['Embarked'] = df_train['Embarked'].map(binary_embarked)
df_test['Embarked'] = df_test['Embarked'].map(binary_embarked)

df_train.head()
```

```python
train = df_train[['Survived', 'Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]
test = df_test[['Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]

train.head()
```

## X, y 데이터로 분리

```python
from sklearn.model_selection import train_test_split

y_train = train['Survived']
X_train = train.drop(['Survived'], axis=1)

X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train)
```

```python
X_test = test
```

## 모델링

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf1 = DecisionTreeClassifier()
dt_clf1.fit(X_train, y_train)

dt_clf2 = DecisionTreeClassifier()
dt_clf2.fit(X_train, y_train)

dt_clf3 = DecisionTreeClassifier()
dt_clf3.fit(X_train, y_train)

print('='*20, 'tree1', '='*20)
print('training set accuracy :', dt_clf1.score(X_train, y_train))
print('valid set accuracy :', dt_clf1.score(X_valid, y_valid))
print('='*20, 'tree2', '='*20)
print('training set accuracy :', dt_clf2.score(X_train, y_train))
print('valid set accuracy :', dt_clf2.score(X_valid, y_valid))
print('='*20, 'tree3', '='*20)
print('training set accuracy :', dt_clf3.score(X_train, y_train))
print('valid set accuracy :', dt_clf3.score(X_valid, y_valid))
```

## 예측

```python
dt_clf1.predict([ [1, 2.0, 0, 100, 0] ])
```

```python
pred = dt_clf1.predict(X_test)
```
