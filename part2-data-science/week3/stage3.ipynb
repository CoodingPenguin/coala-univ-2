```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set()
```

```python
df = pd.read_csv('./data/train.csv')
df.head()
```

## 그래프 그리기

```python
sns.catplot(data=df, x='Age', hue='Survived', aspect=4, kind='count')
```

* 아래 x축 값들이 잘 보이지 않음.
* 너무 촘촘해서 분석하기가 어려움.
* 0.3, 0.4 데이터가 천차만별임.

```python
facet = sns.FacetGrid(df, hue='Survived', aspect=4)
facet.map(sns.kdeplot, 'Age')
facet.add_legend()

plt.show()
```

```python
sns.catplot(data=df, x='Embarked', y='Age', hue='Sex', kind='box', aspect=4)
```

```python
sns.catplot(data=df, x='Embarked', y='Age', hue='Sex', kind='violin', aspect=4)
```
