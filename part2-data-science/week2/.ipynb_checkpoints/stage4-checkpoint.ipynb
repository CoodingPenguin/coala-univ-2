# Stage 04: 시각화 연습하기

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

## 1. matplotlib.pyplot

```python
plt.plot(['Math', 'Korean', 'English'], [90, 100, 50])
plt.plot(['Math', 'Korean', 'English'], [100, 60, 70])
plt.ylabel('Score')
plt.xlabel('Subject')

plt.legend(['Koala', 'Kangaroo'])
plt.show()
```

```python
plt.plot([1, 2, 3], [90, 100, 50], marker='o', markersize=5, linestyle='--')

plt.grid(True)
plt.axis([0, 5, 0, 150])
plt.show()
```

```python
x = ['Math', 'Korean', 'English']
y = [90, 100, 50]

plt.bar(x, y, width=0.3, color='pink')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.show()
```

```python
plt.plot([1, 2, 3], [30, 40, 30], color='black', linestyle='-', linewidth=0.7)
plt.plot([7, 8, 9], [30, 40, 30], color='black', linestyle='-', linewidth=0.7)
plt.plot([2, 8], [30, 30], marker='o', markersize=30, linestyle='', color='black')
plt.plot([4, 5, 6], [20, 10, 20], color='red', linestyle='-')

plt.show()
```

## 2. seaborn

```python
import pandas as pd 

df = pd.read_csv('./data/train.csv')
df.head()
```

```python
sns.set()
```

```python
sns.relplot(data=df, x='Age', y='Survived', aspect=3, hue='Sex')
plt.show()
```

```python
sns.catplot(data=df, x='Pclass', y='Fare', aspect=2, jitter=True)    # jitter로 어디에 많이 분포했는지 알 수 있음
plt.show()
```

```python
sns.pairplot(data=df, hue='Sex',
            x_vars=['Pclass', 'Age', 'Fare'], 
            y_vars=['Pclass', 'Age', 'Fare'],
            height=3)
plt.show()
```

```python
sns.catplot(data=df, x='Pclass', y='Fare', kind='box', aspect=2)
plt.show()
```
